```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# loading data from csv file and create pandas dataframe
df = pd.read_csv('../data/01_raw/data_secop.csv', low_memory=False)
```

```python
# create dataframe for unique values from each column
data = {}
for column in df.columns:
    data[column] = [[df[column].unique()], len(df[column].unique())]
uniques = pd.DataFrame(data)
```

```python
uniques.T
```

```python
df.nunique()
```

```python
# create pandas dataframe from null values
nulls = {}

for column in df.columns:
    nulls[column] = df[column].isnull().sum()

nulls = pd.DataFrame(nulls, index=['count'])
nulls.T.sort_values('count', ascending=False)
```

```python
# total nulls
print("Total nulls in dataframe: ", df.isnull().sum().sum())
```

```python
# Create 5x10 grid of subplots
fig, axs = plt.subplots(6, 10, figsize=(15,20))

axes = axs.flatten()
for ax, column in zip(axes[:-1], df.columns):
  ax.set_title(column, fontdict={"fontsize": "small"})
  ax.set_ylabel('Count')
  ax.hist(df[column].value_counts())

plt.tight_layout()
plt.show()
```

### data visualization

```python
plt.scatter(df["Entidad"].value_counts().index, df["Entidad"].value_counts().values)
plt.title("Scatter Plot Entidad")
plt.xlabel("Entidades") 
plt.ylabel("Conteos")
plt.show()
```
