```python
# read secrets to authentication
import yaml
with open('../conf/local/credentials.yml', 'r') as file:
    credentials = yaml.safe_load(file)
    secop_app_token = credentials['secop_app_token']
```

```python
from sodapy import Socrata
import pandas as pd
import numpy as np

# create socrata client
client = Socrata(domain="www.datos.gov.co",
                 app_token=secop_app_token,
                 )
# query resource to endpoint
results = client.get("p6dx-8zbt", limit=10)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```
